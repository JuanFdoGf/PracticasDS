```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
```

```python
#Cargar el dataset
df = pd.read_csv('churn.csv')

#Verificar las columnas disponibles
print(df.columns)

#Verificar valores nulos
print(df.isnull().sum())

#Eliminar filas duplicadas
df = df.drop_duplicates()

#Verificar las primeras filas
print(df.head())
```

```python
#Informaci√≥n basica del dataframe
print(df.info())

#Resumen estadistico basico
print(df.describe())
```

```python
#Distribuci√≥n de la variable objetivo
sns.countplot(x='Exited', data=df)
plt.title('Distibuci√≥n de Exited (Perdida de Clientes)')
plt.show()
```

```python
#An√°lisis Univariado
df.hist(bins=20, figsize=(15,10))
plt.show()
```

```python
#An√°lisis Bivariado
sns.boxplot(x='Exited', y='Balance', data=df)
plt.title('Distribuci√≥n de Balance vs Exited')
plt.show()
```

## - No se observa una diferencia significativa en la mediana o en el rango de los saldos entre los clientes que abandonaron y los que no. Esto indica que el saldo no parece ser un factor determinante para el abandono de clientes en este caso.
## - La similitud en la distribuci√≥n de saldos sugiere que el saldo de la cuenta podr√≠a no estar directamente relacionado con la probabilidad de que un cliente abandone o no.

```python
sns.boxplot(x='Exited', y='CreditScore', data=df)
plt.title('Distribuci√≥n de CreditScore vs Exited')
plt.show()
```

## - Los clientes que abandonaron la compa√±√≠a tienden a tener un puntaje de cr√©dito ligeramente m√°s bajo en comparaci√≥n con los clientes que permanecieron.
## - La presencia de valores at√≠picos en el grupo de clientes que abandonaron indica que algunos de estos clientes tienen puntajes de cr√©dito notablemente bajos, lo cual podr√≠a contribuir a su salida de la compa√±√≠a.

```python
#Agrupar por pa√≠s y calcular el salario promedio
df_geography_salary = df.groupby('Geography')['EstimatedSalary'].mean()

#Visualizaci√≥n de los salarios por pa√≠s
plt.figure(figsize=(10,6))
ax = sns.barplot(x='Geography', y='EstimatedSalary',
           data=pd.DataFrame({'Geography': df_geography_salary.index, 
                              'EstimatedSalary': df_geography_salary.values}),
           hue='Geography', palette='Greens_d', legend=False)

for i, value in enumerate(df_geography_salary):
    ax.text(i, value + 50, f'${value:.0f}', 
            ha='center', 
            va='bottom', 
            fontweight='bold')
    
plt.title('Salario Promedio por Pa√≠s', pad=20, fontsize=14, fontweight='bold')
plt.ylabel('Salario Estimado ($)', fontsize=12)
plt.xlabel('Pa√≠s', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```
